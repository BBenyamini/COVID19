```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
import numpy as np
import matplotlib.dates as mdates
```

```python
# source of the data
#read the data direwctly from the web
url="https://covid.ourworldindata.org/data/ecdc/full_data.csv"
cov19=pd.read_csv(url,delimiter=",",parse_dates=['date'])
cov19.head()
```

```python
# select data for Rwanda only 
rw_cov=cov19[cov19["location"]=="Rwanda"]
rw_cov.head()
```

```python
# Now we can plot new and total cases for Rwanda

plt.figure(figsize=(20,5))

plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
plt.gca().xaxis.set_major_locator(mdates.DayLocator(interval=3))

plt.title("COVID-19: Cases in Rwanda", fontsize=15)
plt.plot( 'date', 'new_cases', data=rw_cov, color='olive', linewidth=2)
plt.plot( 'date', 'biweekly_cases', data=rw_cov, color='blue', linewidth=2)
plt.plot( 'date', 'total_cases', data=rw_cov, marker='', color='chocolate', linewidth=2)
plt.xticks(rotation=90)

plt.legend()
plt.savefig('Rw_covid19')
```
